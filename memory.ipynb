# Alibaba Cloud Server Memory Optimization

Data

```python
import pandas as pd
df  = pd.read_csv("data.csv")
dataset = df.values
X = dataset[:, 2] # cpu column
Y = dataset[:, 3] # memory column
ID = dataset[:, 1] # node id column
```

Plotting the Data

```python
import plotly.express as px
fig = px.scatter(X, Y, labels={"x" : "Memory Usage"})
fig.show()
```

Model - Multi Layer Perception Neural Network

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
# reshaping the X-factor from 1D to 2D
X = X.reshape(-1, 1)
# splitting the data to train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=1)
# neural network configuration
model = MLPRegressor(random_state=1, max_iter=200, activation="relu", solver="sgd")
# relu = rectifier linear unit is an neural network activation function
# solver = optimization method
# sgd = stochastic gradient descent is an iterative method for optimizing an objective function
model.fit(X_train, Y_train)
# performing optimization on the test set and storing the predicted/optimized memory usage
Y_predict = model.predict(X_test)
```

Plotting the Optimized Memory Usage on Test Data

```python
fig2 = px.scatter(X_test, Y_predict, labels={"x" : "Optimized Memory Usage"})
fig2.show()
```

Node ID and their Optimized Memory Usage

```python
listY_test = list(Y_test)
listID = list(ID)
listY_predict = list(Y_predict)
tmp = []
# list of node id's in the test set
for i in listY_test:
    tmp.append(listY_test.index(i))
listY_test_id = []
for i in tmp:
    listY_test_id.append(listID[i])
    
# printing node id's and their optimized usage
for i, j in zip(listY_test_id, listY_predict):
    print("Node ID =", i, "Optimized Memory Usage =", j)
```
